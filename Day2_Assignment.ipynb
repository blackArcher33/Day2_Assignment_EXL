## Question 1: Data Quality Assessment
1. Load the dataset.
2. Calculate missing values for each column and overall missing percentage.
3. Identify duplicate rows and duplicate transaction groups.
4. Report data types, memory usage, and any unusual data characteristics.

```python
import pandas as pd
df = pd.read_csv('ecommerce_transactions.csv')
df.head()
```

```python
print(df.shape)
```

```python
# Missing values per column
missing_values = df.isnull().sum()

# Missing percentage per column
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_report = pd.DataFrame({
    'Missing_Count': missing_values,
    'Missing_Percentage': missing_percentage
})

print("\nMissing Value Report:")
print(missing_report)

# Overall Missing Percentage
total_missing = df.isnull().sum().sum()
total_cells = df.shape[0] * df.shape[1]

overall_missing_percentage = (total_missing / total_cells) * 100

print("\nOverall Missing Percentage:",
      round(overall_missing_percentage, 2), "%")
```

```python
duplicate_rows = df[df.duplicated()]

print("\nNumber of Duplicate Rows:",
      duplicate_rows.shape[0])

print("\nDuplicate Rows:")
print(duplicate_rows)
```

```python
duplicate_transaction_groups = df[
    df.duplicated(subset=['Transaction_ID'], keep=False)
]

print("\nDuplicate Transaction Groups:")
print(duplicate_transaction_groups)

print("\nNumber of Duplicate Transaction Records:",
      duplicate_transaction_groups.shape[0])
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
print("\nMemory Usage by Column (Bytes):")
print(df.memory_usage(deep=True))

print("\nTotal Memory Usage (MB):")
print(round(df.memory_usage(deep=True).sum() / (1024**2), 4))
```

## Question 3: Outlier Detection
1. Use the IQR method to flag outliers for numeric columns.
2. Use the Z-score method to flag outliers for numeric columns.
3. Compare both methods and list the top 10 `final_price` outliers.

```python
# Select Numeric Columns
numeric_cols = df.select_dtypes(include=['int64', 'float64'])

def detect_iqr_outliers(data, column):

    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[
        (data[column] < lower_bound) |
        (data[column] > upper_bound)
    ]

    return outliers
```

```python
for col in numeric_cols.columns:

    outlier_rows = detect_iqr_outliers(df, col)

    print("\nColumn:", col)
    print("Number of Outliers:", len(outlier_rows))
```

```python
#z Score Function
from scipy.stats import zscore
def detect_zscore_outliers(data, column):

    z_scores = zscore(data[column])

    outliers = data[abs(z_scores) > 3]

    return outliers
```

```python
for col in numeric_cols.columns:

    outlier_rows = detect_zscore_outliers(df, col)

    print("\nColumn:", col)
    print("Number of Outliers:", len(outlier_rows))
```

```python
comparison = []

for col in numeric_cols.columns:

    iqr_count = len(detect_iqr_outliers(df, col))
    zscore_count = len(detect_zscore_outliers(df, col))

    comparison.append([
        col,
        iqr_count,
        zscore_count
    ])

comparison_df = pd.DataFrame(
    comparison,
    columns=[
        "Column",
        "IQR_Outliers",
        "ZScore_Outliers"
    ]
)

print(comparison_df)
```

```python
Q1 = df['Purchase_Amount'].quantile(0.25)
Q3 = df['Purchase_Amount'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

final_price_outliers = df[
    (df['Purchase_Amount'] < lower) |
    (df['Purchase_Amount'] > upper)
]

top10_outliers = final_price_outliers.sort_values(
    by='Purchase_Amount',
    ascending=False
).head(10)

print(top10_outliers)
```
